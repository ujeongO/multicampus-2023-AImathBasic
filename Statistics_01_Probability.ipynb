```python
from scipy import stats
import seaborn as sns
import numpy as np
import matplotlib.pylab as plt

%matplotlib inline

n = 10      # 전체 시도 횟수
mu = 0.6    # 베르누이 확률분포 기대값

rv = stats.binom(n, mu)
# binom: binominal
xx = np.arange(n+1)
# n+1 >> n번까지

plt.bar(xx, rv.pmf(xx), align='center')
# pmf(Probability Mass Function): 확률질량함수(=확률밀도함수)
# align: 정렬

plt.xlabel("표본값")
plt.ylabel("$P(x)$")
plt.title("확률질량함수_pmf")

plt.show()

```

```python
# 기대값이 1이고 표준편차가 2인 정규분포 객체 생성

rv = stats.norm()
rv = stats.norm(loc=1, scale=2)
# norm: normal distribution(정규분포)

x = np.linspace(-8, 8, 100)
# -8 ~ 8 사이를 100등분 해줘
pdf = rv.pdf(x)
plt.plot(x, pdf)

plt.show()
```

확률분포의 유형_정규분포(Normal distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 평균, 표준편차

mu, sigma = 0, 0.1
```

```python
np.random.normal(mu, sigma, 10)
```

```python
s = np.random.normal(mu, sigma, 1000)

# 평균 0, 표준편차 0.1, 샘플수 1000
# (참고-이론) 통계적으로 sample 30개 이상 >> prototype
```

```python
# 확률밀도함수(histogram)

# count, bins, ignored = plt.hist(s, 30, normed=True)
# Attribute Error >> normed -> density (옵션 이름 바뀜)
count, bins, ignored = plt.hist(s, 30, density=True)
plt.show()
# 30 >> bin(계급) 
```

```python
# 표준정규분포 공식
plt.plot(bins, 1/(sigma * np.sqrt(2*np.pi)) *
         np.exp(-(bins-mu)**2 / (2 * sigma**2)),
         linewidth=2, color='r'
        )

plt.show()
```

```python
# 두 그래프를 합치면

count, bins, ignored = plt.hist(s, 30, density=True)
# plt.show()

plt.plot(bins, 1/(sigma * np.sqrt(2*np.pi)) *
         np.exp(-(bins-mu)**2 / (2 * sigma**2)),
         linewidth=2, color='r'
        )

plt.show()
```
